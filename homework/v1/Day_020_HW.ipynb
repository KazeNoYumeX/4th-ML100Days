```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 另一個繪圖-樣式套件
import seaborn as sns

plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings

warnings.filterwarnings('ignore')
```

```python
# Create a 10x10 matrix with random values from -1 to 1
matrix = np.random.uniform(-1, 1, (10, 10))
plt.figure(figsize=(10, 10))

# Print the heatmap
cmap = mpl.colormaps['RdYlBu_r']
heatmap = sns.heatmap(matrix, cmap=cmap, annot=True)
plt.show()
```

```python
# Create a 1000x3 matrix with random values from -1 to 1
n_row = 1000
n_col = 3
matrix = np.random.uniform(-1, 1, (n_row, n_col))

# Randomly assign 0, 1, 2 to the labels
indices = np.random.choice([0, 1, 2], size=n_row)
plot_data = pd.DataFrame(matrix, indices).reset_index()

# 繪製 seaborn 進階 Heatmap
grid = sns.PairGrid(data=plot_data, height=3, diag_sharey=False,
                    hue='index', vars=[x for x in list(plot_data.columns) if x != 'index'])

# Scatter plot on the upper part
grid.map_upper(plt.scatter, alpha=0.2)

# Histogram on the diagonal
grid.map_diag(sns.kdeplot)

# Density plot on the lower part
cmap = mpl.colormaps['OrRd_r']
grid.map_lower(sns.kdeplot, cmap=cmap)
plt.show()
```

```python
# Create a 1000x3 matrix with randomness from normal distribution
n_row = 1000
n_col = 3

matrix = np.random.randn(n_row, n_col)

# Randomly assign 0, 1, 2 to the labels
indices = np.random.choice([0, 1, 2], size=n_row)
plot_data = pd.DataFrame(matrix, indices).reset_index()

# 繪製 seaborn 進階 Heatmap
grid = sns.PairGrid(data=plot_data, height=3, diag_sharey=False,
                    hue='index', vars=[x for x in list(plot_data.columns) if x != 'index'])

# Scatter plot on the upper part
grid.map_upper(plt.scatter, alpha=0.2)

# Histogram on the diagonal
grid.map_diag(sns.kdeplot)

# Density plot on the lower part
cmap = mpl.colormaps['OrRd_r']
grid.map_lower(sns.kdeplot, cmap=cmap)

plt.show()
```
