```python
import pandas as pd
import seaborn as sns
from sklearn import datasets

data = datasets.load_breast_cancer()
cancer = pd.DataFrame(data["data"], columns=data["feature_names"])
cancer["target"] = data["target"]
```

```python
# 資料視覺化
sns.relplot(x="mean radius", y="mean texture", hue="target", data=cancer)
```

```python
from sklearn.linear_model import LogisticRegression

# 訓練模型
X = cancer.iloc[:, 0:30].values
y = cancer["target"].values

# Because the lbfgs failed to converge, we change the solver to liblinear and increase the max_iter to 1000
model = LogisticRegression(solver='liblinear', max_iter=1000)
model.fit(X=X, y=y)
```

```python
# 模型評估
model.score(X=X, y=y)
```

```python
# 模型預測
predict_y = model.predict(X=X)
predict_y
```
