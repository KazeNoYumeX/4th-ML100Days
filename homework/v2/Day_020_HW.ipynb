# [作業]

- 請用最小平方法推導多元迴歸（multiple regression）的公式解（請使用向量表示）。

##### Ans:

## 多元線性迴歸如何用矩陣方式呈現？
- 已知每筆資料每個特徵的多元線性迴歸方程式如下

$$
\begin{cases}
y_{1}=b_0+b_1\times x_{11}+b_2\times x_{12}+\cdots+b_p\times x_{1p} \\
y_{2}=b_0+b_1\times x_{21}+b_2\times x_{22}+\cdots+b_p\times x_{2p} \\
\vdots  \\
y_{n}=b_0+b_1\times x_{n1}+b_2\times x_{n2}+\cdots+b_p\times x_{np}
\end{cases}
$$

- 我們可將上述方程式以矩陣方式表達。
    - **Y**：目標數值矩陣，大小為 $(n\times 1)$
    - **X**：特徵矩陣，大小為 $(n\times p)$
    - $\bf{b}$：迴歸係數矩陣，大小為 $(p\times 1)$

$$
\bf{Y}=\bf{X}\bf{b}+\epsilon
$$

$$
\bf{Y}=
\begin{pmatrix}
y_1 \\
y_2 \\
\vdots \\
y_n
\end{pmatrix}_{(n\times 1)}
$$

$$
\bf{X}=
\begin{pmatrix}
    x_{11} & x_{12} & \cdots & x_{1p} \\
    x_{21} & x_{22} & \cdots & x_{2p} \\
    \vdots & \vdots & \cdots & \vdots \\
    x_{n1} & x_{n2} & \cdots & x_{np}
\end{pmatrix}_{(n\times p)}
$$

$$
\bf{b}=
\begin{pmatrix}
b_0 \\
b_1 \\
\vdots \\
b_p
\end{pmatrix}_{(p\times 1)}
$$


## 如何用最小平方法證明 $\bf{b}$ 公式解？
- 已知 $\bf{Y}=\{y_1, y_2, \cdots, y_n\}$ 為所有資料的真實答案、$\bf{\hat{Y}}=\{\hat{y}_1, \hat{y}_2, \cdots, \hat{y}_n\}$
- Step1. 用矩陣方式計算殘差 $e$

$$
L = e^2=
(\bf{Y}-\bf{\hat{Y}})^2=
(Y-\bf{X}\bf{b})^2=
(\bf{Y}-\bf{X}\bf{b})^T(\bf{Y}-\bf{X}\bf{b})
$$

- Step2. 將公式拆解成易於偏微分的狀態

$$
L =
(\bf{Y}-\bf{X}\bf{b})^T(\bf{Y}-\bf{X}\bf{b}) \\=
(Y^T-\bf{b}^T\bf{X}^T)(\bf{Y}-\bf{X}\bf{b}) \\=
Y^TY-Y^T\bf{X}\bf{b}-\bf{b}^T\bf{X}^T\bf{Y}+\bf{b}^T\bf{X}^T\bf{X}\bf{b} \\
=\bf{Y}^T\bf{Y}-2\bf{b}^T\bf{X}^TY+\bf{b}^T\bf{X}^T\bf{X}\bf{b}
$$

- Step3. 將 $\bf{b}$ 對 $L$ 做偏微分，並讓答案為零

$$
\frac{\partial L}{\partial \bf{b}}=-2\bf{X}^T\bf{Y}+2\bf{X}^T\bf{X}\bf{b}=0
$$

- Step4. 求解 $\bf{b}$

$$
\bf{X}^T\bf{X}\bf{b}=\bf{X}^T\bf{Y} \\
\bf{b}=(\bf{X}^T\bf{X})^{-1}\bf{X}^T\bf{Y}
$$
